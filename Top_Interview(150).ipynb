```python
#https://leetcode.com/problems/merge-sorted-array/
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        i = m - 1        # last valid element in nums1
        j = n - 1        # last element in nums2
        k = m + n - 1    # last position in nums1

        while i >= 0 and j >= 0:
            if nums1[i] > nums2[j]:
                nums1[k] = nums1[i]
                i -= 1
            else:
                nums1[k] = nums2[j]
                j -= 1
            k -= 1

        while j >= 0:
            nums1[k] = nums2[j]
            j -= 1
            k -= 1
```

```python
#https://leetcode.com/problems/remove-element/
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        left = 0
        for right in range(len(nums)):
            if nums[right] != val:
                nums[left] = nums[right]
                left += 1
        return left
```

```python
#https://leetcode.com/problems/merge-two-sorted-lists/
class Solution:
    def mergeTwoLists(
        self,
        list1: Optional[ListNode],   # head of first sorted linked list
        list2: Optional[ListNode]    # head of second sorted linked list
    ) -> Optional[ListNode]:

        dummy = ListNode(-1)
        # Dummy node acts as a fixed starting point.
        # It simplifies edge cases (like empty lists).

        current = dummy
        # Pointer that will build the merged list step by step.
        # It always points to the last node in the merged list.

        while list1 and list2:
            # Continue while both lists still have nodes

            if list1.val <= list2.val:
                # If current node in list1 has a smaller (or equal) value

                current.next = list1
                # Attach list1's current node to the merged list

                list1 = list1.next
                # Move forward in list1

            else:
                # If current node in list2 has a smaller value

                current.next = list2
                # Attach list2's current node to the merged list

                list2 = list2.next
                # Move forward in list2

            current = current.next
            # Move the merged-list pointer forward

        # At this point, at least one list is empty

        current.next = list1 if list1 else list2
        # Attach the remaining nodes (only one of these is non-empty)

        return dummy.next
        # Return the head of the merged list (skip the dummy node)
```

**2516. Take K of Each Character From Left and Right**

The problem requires removing characters only from the left and right ends of the string such that we take at least k of each character ('a', 'b', 'c') while minimizing the total number of removed characters.

Instead of deciding which characters to remove, we reverse the perspective and decide which characters we are allowed to keep in the middle. This works because keeping more characters in the middle automatically means removing fewer characters from the ends.

For a character c:

*   The string contains s.count(c) total occurrences.
*   We must remove at least k occurrences of c.
*   Therefore, the maximum number of c that can remain in the middle is s.count(c) - k.

```python
#https://leetcode.com/problems/take-k-of-each-character-from-left-and-right/
class Solution:
    def takeCharacters(self, s: str, k: int) -> int:
        # Count how many characters we are allowed to KEEP in the middle
        allowed = {}
        for c in "abc":
            allowed[c] = s.count(c) - k

        # If we can't remove k of each character, it's impossible
        if min(allowed.values()) < 0:
            return -1

        # Sliding window to find the longest substring we can keep
        window_count = {"a": 0, "b": 0, "c": 0}
        left = 0
        max_window = 0

        #left is the index of the first character in your current window
        #right is the index of the last character in your current window
        for right in range(len(s)):
            char = s[right]
            window_count[char] += 1

            # If the count of the current character exceeds allowed[char] (i.e., we would be keeping too many of this character), we shrink the window from the left.
            while window_count[char] > allowed[char]:
                window_count[s[left]] -= 1 # Subtracting 1 from the count of s[left] (removing the character at the start of the window)
                left += 1

            # Update the maximum valid window size
            max_window = max(max_window, right - left + 1)

        # Total length minus the kept window = characters taken from ends
        return len(s) - max_window
```

```python
#https://leetcode.com/problems/maximize-area-of-square-hole-in-grid/
class Solution:
    def maximizeSquareHoleArea(self, n: int, m: int, hBars: List[int], vBars: List[int]) -> int:
        h = len(hBars)
        v = len(vBars)
        hBars.sort()
        vBars.sort()

        def find_continuous_chunk(nums) -> list:
            if not nums:
                print("warning: empty inputs")
                return []
            counts = 1
            res = 0
            n = len(nums)
            for i in range(n-1):
                if nums[i] == nums[i+1] - 1:
                    counts += 1
                    res = max(res, counts)
                else:
                    counts = 1
            return max(res, counts)

        hcounts = find_continuous_chunk(hBars)
        vcounts = find_continuous_chunk(vBars)
        return (min(vcounts, hcounts)+1) ** 2
```

*   Finding the size of a matrix using two variables.
*   Building a prefix sum matrix initialized with zeros.
*   Filling the prefix sum matrix with cumulative sums from the original matrix.
*   Checking whether any **k Ã— k** submatrix has a sum less than or equal to the threshold.
*   Finding the largest square size whose sum is within the threshold.







```python
#https://leetcode.com/problems/maximum-side-length-of-a-square-with-sum-less-than-or-equal-to-threshold/
from typing import List

class Solution:
    def maxSideLength(self, mat: List[List[int]], threshold: int) -> int:
        m, n = len(mat), len(mat[0])

        # build prefix sum matrix
        pre = []
        for i in range(m + 1):
          row = []
          for j in range(n + 1):
            row.append(0)
          pre.append(row)

        for i in range(m):
            row_sum = 0
            for j in range(n):
                row_sum += mat[i][j]
                pre[i + 1][j + 1] = pre[i][j + 1] + row_sum

        def square_exists(k: int) -> bool:
            # check if any k x k square has sum <= threshold
            for i in range(k, m + 1):
                for j in range(k, n + 1):
                    s = (pre[i][j] - pre[i - k][j] - pre[i][j - k] + pre[i - k][j - k])
                    if s <= threshold:
                        return True
            return False

        ans = 0
        # grow side length while feasible
        for k in range(1, min(m, n) + 1):
            if square_exists(k):
                ans = k
            else:
                break
        return ans
```
